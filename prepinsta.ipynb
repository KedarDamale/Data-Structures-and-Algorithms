### Getting Started


Positive or Negative number

```python
n=int(input("Enter a number to find out if it is positive or negative"))
def positive_or_negative(n):
    if n>=0:
        return "Positive"
    else:
        return "Negative"

positive_or_negative(n)
```

Even or odd number

```python
n=int(input("Enter a number to find out if it is even or odd"))
def even_or_odd(n):
    if n%2==0:
        return "Even"
    else:
        return "Odd"

even_or_odd(n)
```

Sum of First N Natural numbers

```python
#using for loop

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_using_for_loop(n):
    sum=0

    for i in range(1,n+1):
        sum+=i

    return sum

sum_of_n_using_for_loop(n)
```
