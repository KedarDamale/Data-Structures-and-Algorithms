Positive or Negative number

```python
n=int(input("Enter a number to find out if it is positive or negative"))
def positive_or_negative(n):
    if n>=0:
        return "Positive"
    else:
        return "Negative"

positive_or_negative(n)
```

Even or odd number

```python
n=int(input("Enter a number to find out if it is even or odd"))
def even_or_odd(n):
    if n%2==0:
        return "Even"
    else:
        return "Odd"

even_or_odd(n)
```

Sum of First N Natural numbers

```python
#using for loop

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_using_for_loop(n):
    sum=0

    for i in range(1,n+1):
        sum+=i

    return sum

sum_of_n_using_for_loop(n)
```

```python
#using formula

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_formula(n):
    return n*(n+1)/2

sum_of_n_formula(n)
```

```python
#using recursion

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n(n):

    if n==1:
        return 1
    else:
        return n+sum_of_n(n-1)

sum_of_n(n) 
```

Find the Sum of Numbers in a given Range in Python


```python
#using loop

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_of_n_in_range(m,n):
    sum=0
    for i in range(m,n+1):
        sum+=i
    return sum
sum_of_n_in_range(m,n)
```

```python
#using recursion

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_using_recursion(m,n):
    if m==n:
        return m
    else:
        return n+sum_using_recursion(m,n-1)

sum_using_recursion(m,n)
```

Find the Greatest of Three Numbers in Python


```python
#using max function is basically max(a,b,c)

#using if else

            #     a>b
            #     / \ 
            #    T   F
            #   a>c  b>c
            # /    \ /   \ 
            # T     F T   F 
            # a     c b    c


def max_between_three(a,b,c):

    if a>b:
        if a>c:
            print(a)
        else:
            print(c)
    else:
        if b>c:
            print(b)
        else:
            print(c)

max_between_three(89,45,67)
```

Prime Number

```python
#without optimization

num = 15
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

```python
#square root optimization

num = 7
flag = 0
if num<2:
  flag = 1
else:
  for i in range(2,int(pow(num,0.5)+1)):
    if num%i==0:
      flag = 1
      break

if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

Sum of Digits of a Number in Python


```python
#extracting digits and then sum
n=1678
sum=0
while n>0:
    digit=n%10
    n//=10
    sum+=digit
print(sum)
```

Reverse a Number

```python
num = 1234
temp = num
reverse = 0
while num > 0:
    remainder = num % 10
    reverse = (reverse * 10) + remainder
    num = num // 10

print(reverse)
```
