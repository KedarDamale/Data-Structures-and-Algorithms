Sum of First N Natural numbers

```python
#using for loop

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_using_for_loop(n):
    sum=0

    for i in range(1,n+1):
        sum+=i

    return sum

sum_of_n_using_for_loop(n)
```

```python
#using formula

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_formula(n):
    return n*(n+1)/2

sum_of_n_formula(n)
```

```python
#using recursion

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n(n):

    if n==1:
        return 1
    else:
        return n+sum_of_n(n-1)

sum_of_n(n) 
```

Find the Sum of Numbers in a given Range in Python


```python
#using loop

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_of_n_in_range(m,n):
    sum=0
    for i in range(m,n+1):
        sum+=i
    return sum
sum_of_n_in_range(m,n)
```

```python
#using recursion

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_using_recursion(m,n):
    if m==n:
        return m
    else:
        return n+sum_using_recursion(m,n-1)

sum_using_recursion(m,n)
```

Find the Greatest of Three Numbers in Python


```python
#using max function is basically max(a,b,c)

#using if else

            #     a>b
            #     / \ 
            #    T   F
            #   a>c  b>c
            # /    \ /   \ 
            # T     F T   F 
            # a     c b    c


def max_between_three(a,b,c):

    if a>b:
        if a>c:
            print(a)
        else:
            print(c)
    else:
        if b>c:
            print(b)
        else:
            print(c)

max_between_three(89,45,67)
```

Prime Number

```python
#without optimization

num = 15
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

```python
#square root optimization

num = 7
flag = 0
if num<2:
  flag = 1
else:
  for i in range(2,int(pow(num,0.5)+1)):
    if num%i==0:
      flag = 1
      break

if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

Armstrong Number

```python
# sum of powers of each digits should equate to the numbebr where power is count of digits
# 1634=1^4+6^4+3^4+4^4

n=370

def armstrong(n):
    temp=n
    count=len(str(n))
    am=0
    while n>0:
        digit=n%10
        am+=(digit**count)
        n//=10
    if temp==am:
        return True
    return False


armstrong(n)
```

Armstrong Number in a range

```python
m=0
n=100

for i in range(m,n+1):
    if armstrong(i):
        print(i)
```

Fibonacci Series

```python
#usign recursion

fib_series=[0,1]
def fibo(n):
    if n==0:
        return 0
    if n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)

fibo(7)
```

factor pairs of a number


```python
def factors(n):
    factors=[(1,n)]
    for i in range(2,n):
        if n%i==0:
            factors.append((i,(n//i)))
    
    factors.append((n,1))
    return factors

factors(100)
```

A strong number (or factorion) is a positive integer that equals the sum of the factorials of its individual digits

```python
n=145
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
def strong_number(n):
    temp=n
    total_sum=0
    while n>0:
        digit=n%10
        total_sum+=fact(digit)
        n=n//10
    if temp==total_sum:
        return True
    else: 
        return False

strong_number(n)
```

Highest Common Factor(HCF)

```python
#6 is the highest common factor of 12 and 18

def hcf(a,b):
    curr_hcf=1
    for i in range(1,min(a,b)+1):
        if a%i==0 and b%i==0:
            if i>curr_hcf:
                curr_hcf=i

    return curr_hcf

hcf(12,18)
```

Lowest Common Multiple (LCM)

```python
#The Lowest Common Multiple (LCM) is the smallest positive whole number that is a multiple of two or more given numbers
#LCM(4, 6): Multiples of 4 (4, 8, 12, 16...) and 6 (6, 12, 18...) -> LCM is 12.

def lcm(a,b):
    return a*b/hcf(a,b)

lcm(4,6)
```
