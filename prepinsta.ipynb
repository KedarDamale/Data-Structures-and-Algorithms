### Getting Started


Positive or Negative number

```python
n=int(input("Enter a number to find out if it is positive or negative"))
def positive_or_negative(n):
    if n>=0:
        return "Positive"
    else:
        return "Negative"

positive_or_negative(n)
```

Even or odd number

```python
n=int(input("Enter a number to find out if it is even or odd"))
def even_or_odd(n):
    if n%2==0:
        return "Even"
    else:
        return "Odd"

even_or_odd(n)
```

Sum of First N Natural numbers

```python
#using for loop

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_using_for_loop(n):
    sum=0

    for i in range(1,n+1):
        sum+=i

    return sum

sum_of_n_using_for_loop(n)
```

```python
#using formula

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n_formula(n):
    return n*(n+1)/2

sum_of_n_formula(n)
```

```python
#using recursion

n=int(input("Enter a number to find out sum of first n numbers"))

def sum_of_n(n):

    if n==1:
        return 1
    else:
        return n+sum_of_n(n-1)

sum_of_n(n) 
```

Find the Sum of Numbers in a given Range in Python


```python
#using loop

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_of_n_in_range(m,n):
    sum=0
    for i in range(m,n+1):
        sum+=i
    return sum
sum_of_n_in_range(m,n)
```

```python
#using recursion

m,n=input("Enter range separated by space").split()
m=int(m)
n=int(n)

def sum_using_recursion(m,n):
    if m==n:
        return m
    else:
        return n+sum_using_recursion(m,n-1)

sum_using_recursion(m,n)
```
